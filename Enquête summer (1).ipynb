```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Partie A: Résumé de l'enquête
summary = """
L'enquête SUMER 2017 vise à cartographier les expositions aux risques professionnels
chez les salariés en France, tant dans la fonction publique que dans le secteur privé.
Elle analyse différents types de contraintes (physiques, organisationnelles, biologiques,
chimiques, psychosociales) pour fournir une vue d'ensemble sur les conditions de travail
en termes de risques et protections associées.
"""
print("Résumé de l'enquête :")
print(summary)
```

```python
# Partie B: Génération des tables de données
# Table 1 : Profil des salariés par secteur
profil_data = {
    "Secteur": ["Public - État", "Public - Territorial", "Public - Hospitalier", "Privé"],
    "Effectif": [2070300, 1981200, 1060000, 18000000],
    "Temps partiel (%)": [11.9, 23.8, 27.4, 19.6],
    "Moyenne d'âge": [44, 43, 41, 42],
    "Pourcentage de Femmes (%)": [52.1, 61.0, 77.0, 46.0]
}
profil_df = pd.DataFrame(profil_data)

# Affichage de la table profil des salariés
print("Profil des salariés :")
print(profil_df)

# Table 2 : Expositions par secteur
exposition_data = {
    "Secteur": ["Public - État", "Public - Territorial", "Public - Hospitalier", "Privé"],
    "Contraintes physiques (%)": [55.8, 80.6, 65.2, 71.5],
    "Contraintes organisationnelles (%)": [79.8, 84.5, 88.0, 74.0],
    "Exposition aux agents biologiques (%)": [26.6, 43.0, 70.0, 24.9],
    "Exposition aux agents chimiques (%)": [14.6, 38.5, 24.3, 32.3],
    "Risques psychosociaux (%)": [61.1, 70.3, 75.0, 50.6]
}
exposition_df = pd.DataFrame(exposition_data)

# Affichage de la table expositions
print("Expositions par secteur :")
print(exposition_df)
```

```python

# Enregistrer les tables dans des fichiers CSV
profil_df.to_csv("profil_salaries.csv", index=False)
exposition_df.to_csv("exposition_secteurs.csv", index=False)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Données : Profil des salariés
profil_data = {
    "Secteur": ["Public - État", "Public - Territorial", "Public - Hospitalier", "Privé"],
    "Effectif": [2070300, 1981200, 1060000, 18000000],
    "Temps partiel (%)": [11.9, 23.8, 27.4, 19.6],
    "Moyenne d'âge": [44, 43, 41, 42],
    "Pourcentage de Femmes (%)": [52.1, 61.0, 77.0, 46.0]
}
profil_df = pd.DataFrame(profil_data)

# Données : Expositions par secteur
exposition_data = {
    "Secteur": ["Public - État", "Public - Territorial", "Public - Hospitalier", "Privé"],
    "Contraintes physiques (%)": [55.8, 80.6, 65.2, 71.5],
    "Contraintes organisationnelles (%)": [79.8, 84.5, 88.0, 74.0],
    "Exposition aux agents biologiques (%)": [26.6, 43.0, 70.0, 24.9],
    "Exposition aux agents chimiques (%)": [14.6, 38.5, 24.3, 32.3],
    "Risques psychosociaux (%)": [61.1, 70.3, 75.0, 50.6]
}
exposition_df = pd.DataFrame(exposition_data)

# Visualisation pour chaque indicateur de la table Profil des salariés
profil_indicators = ["Effectif", "Temps partiel (%)", "Moyenne d'âge", "Pourcentage de Femmes (%)"]
for indicator in profil_indicators:
    plt.figure(figsize=(8, 5))
    plt.bar(profil_df["Secteur"], profil_df[indicator], color='skyblue')
    plt.title(f"Diagramme en bâtons : {indicator}", fontsize=14)
    plt.xlabel("Secteur", fontsize=12)
    plt.ylabel(indicator, fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Visualisation pour chaque indicateur de la table Expositions
exposition_indicators = [
    "Contraintes physiques (%)", "Contraintes organisationnelles (%)",
    "Exposition aux agents biologiques (%)", "Exposition aux agents chimiques (%)",
    "Risques psychosociaux (%)"
]
for indicator in exposition_indicators:
    plt.figure(figsize=(8, 5))
    plt.bar(exposition_df["Secteur"], exposition_df[indicator], color='orange')
    plt.title(f"Diagramme en bâtons : {indicator}", fontsize=14)
    plt.xlabel("Secteur", fontsize=12)
    plt.ylabel(indicator, fontsize=12)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```
